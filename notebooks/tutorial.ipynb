### Tutorial

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
mean = [-2, -2]
cov = [[1,0], [0,1]]
x1, y1 = np.random.multivariate_normal(mean, cov, 100).T
```

```python
x1[:10], y1[:10]
```

```python
plt.scatter(x1, y1, color="r", label="d1")
```

```python
# 平均 x = 2, y = 2 の正規分布から100個データをサンプリング
mean2 = [2,2]
cov2 = [[1,0], [0,1]]
x2, y2 = np.random.multivariate_normal(mean2, cov2, 100).T
```

```python
plt.scatter(x2, y2, color="b", label="d2")
```

確かに、設定した平均のあたりにデータが集合していることを確認できる。

```python
plt.scatter(x1, y1, color="r", label="d1")
plt.scatter(x2, y2, color="b", marker="x", label="d2")
plt.grid(True)
plt.legend()
plt.show()
```
